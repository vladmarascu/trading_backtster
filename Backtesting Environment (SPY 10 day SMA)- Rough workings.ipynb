L/S Trading signal:

```python
cond = spy['sma10'] < spy['Close']

# Signal column based on cond
spy['cond'] = np.where(cond, 1, -1)

spy.head(10)
```

However, we only want to change the number of shares in the Portfolio when the indicator changes (from -1 to 1 or viceversa), so we will create some functions to account for this:

- As of now, our script will keep buying and selling as long as the cond is True/False, which is not right

```python
spy['signal']=0

rows=len(spy)
for i in range(1,rows):
    if spy['cond'][i]==spy['cond'][i-1]:
        #spy['signal1'][i-1]=spy['signal'][i-1]
        spy['signal'][i]=1
        i=i+1

spy.head(10)
```

```python
for i in range(0,rows):
    if spy['signal'][i]==1:
        spy['signal'][i]=0
        i=i+1
    else:
        spy['signal'][i]=spy['cond'][i]
        i=i+1

spy.head(10)
```

We have the trading signals based on the imposed cond. 

Check returns (simple backtesting):

Create Portfolio: we split it into shares and cash balance, and we can compute the Portfolio Value based on a combination of these

```python
# Set the initial capital
initial_capital= 100000

spy['shares']=0

spy.drop(['cond'],axis=1, inplace=True)

spy['balance']=float(0.0)
spy['balance'][0]=initial_capital
#spy['etf'] = spy['etf'].astype('float')

spy.head(10)
```

Iterate over the shares and balance columns and compute the Portfolio value over the whole time period:

```python
for i in range(1,len(spy)):
    if spy.signal[i]==1:
        spy['shares'][i]=0.1*spy.balance[i-1]/spy.etf[i]
        spy['balance'][i]=0.9*spy.balance[i-1]
        
        i=i+1
    elif spy.signal[i]==-1:
        spy['balance'][i]=spy.balance[i-1]+spy.shares[i-1]*spy.etf[i]
        spy['shares'][i]=0
        
        i=i+1
    
    else:
        spy.balance[i]=spy.balance[i-1]
        spy.shares[i]=spy.shares[i-1]

spy['Portfolio Value']=spy.balance+spy.etf*spy.shares   

spy.head(10)
```
