To compare for trading signals:

```python
spy[['sma10','sp500']].iplot(kind='line')
```

```python
spy.dropna(inplace=True)
spy.head()
```

```python
cond = spy['sma10'] < spy['sp500']

# Signal column based on cond
spy['cond'] = np.where(cond, 1, -1)

spy.head(10)
```

```python
spy['signal']=0

rows=len(spy)
for i in range(1,rows):
    if spy['cond'][i]==spy['cond'][i-1]:
        #spy['signal1'][i-1]=spy['signal'][i-1]
        spy['signal'][i]=1
        i=i+1

spy.head(10)
```

```python
for i in range(0,rows):
    if spy['signal'][i]==1:
        spy['signal'][i]=0
        i=i+1
    else:
        spy['signal'][i]=spy['cond'][i]
        i=i+1

spy.head(10)
```

Check returns (simple backtesting):

```python
# Calculate Returns
spy['returns'] = spy.Close.pct_change()

# Calculate Strategy Returns
spy['strategy_returns'] = spy.returns * spy.signal.shift(1)

# Calculate Cumulative Returns
cumulative_strategy_returns = (spy.strategy_returns + 1).cumprod()

# Plot Strategy Returns
cumulative_strategy_returns.iplot()
```

Create Portfolio:

```python
# Set the initial capital
initial_capital= float(100000.0)

# Positions
spy['positions'] = 100*spy['signal']   
 
spy.head(10)
```

```python
spy['long_amt']=(-1)*(np.where(spy.signal==1, 100*spy.Close,0))
spy['short_amt']=(1)*(np.where(spy.signal==-1, 100*spy.Close,0))
spy.head(10)
```

```python
spy['cash_delta']=spy['long_amt']+spy['short_amt']
spy.head(10)
```

```python
spy['balance']=initial_capital+spy['cash_delta'].cumsum()
spy.head(10)
```
